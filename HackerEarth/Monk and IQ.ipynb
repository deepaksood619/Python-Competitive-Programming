First Line - C P N
where, 
C = denotes number of courses in the college
P = monk and his friends
N = number of students already applied for the course

Second line - N space separated integers Y[i] which denotes the IQ of the ith student. Here, the ith student chooses the ith course.

Third line - P space separated integers X[i] which denotes the IQ of Monk and his friends.

Output - Print P space separated integers in a line which denotes the course number which Monk and his friends have applied for.

Ex 1:
5 4 4
2 8 5 1
9 10 5 1

Output: 5 4 1 3

Ex 2:
10 10 10
2 8 5 1 10 5 9 9 3 5
2 5 1 8 6 5 1 10 1 6

Output: 4 1 9 3 6 10 4 2 9 9

For each course z = x*c will be calculated (c = number of students already enrolled in the course i and x is the sum of IQ of the last two students who enrolled in that course.)

```python
# using list insted of priority queue

# c, p, n = map(int, input().split())
# students_iq = list(map(int, input().split()))
# monks_iq = list(map(int, input().split()))

c, p, n = 10, 10, 10
students_iq = [2,8,5,1,10,5,9,9,3,5]
monks_iq = [2,5,1,8,6,5,1,10,1,6]

# create a list of number of courses
course_students_count = [0]*c
# first item in each list is the z_index and second item is last_student's iq
z_index = [[0 for x in range(2)] for y in range(c)]

# assign all students to courses
for i in range(n):
    z_index[i][0] = students_iq[i]
    z_index[i][1] = students_iq[i]
    course_students_count[i] = 1
    
print(z_index)
print(course_students_count)

print(z_index.index(min(z_index, key = lambda t: t[0])))

for i in range(p):
    min_index = z_index.index(min(z_index, key=lambda t: t[0]))
    print(min_index + 1, end=' ')
    course_students_count[min_index] += 1
    z_index[min_index][0] = (z_index[min_index][1] + monks_iq[i]) * course_students_count[min_index]
    z_index[min_index][1] = monks_iq[i]
    print(z_index, course_students_count)
```

```python
# using priority queue
# problem insert order not maintained
# can use heap with heapify operation

import queue as Q

# c, p, n = map(int, input().split())
# students_iq = list(map(int, input().split()))
# monks_iq = list(map(int, input().split()))

c, p, n = 10, 10, 10
students_iq = [2,8,5,1,10,5,9,9,3,5]
monks_iq = [2,5,1,8,6,5,1,10,1,6]

# create a list of number of courses
course_students_count = [0]*c
# first item in each list is the z_index and second item is last_student's iq and third item is course number
z_index = Q.PriorityQueue()

# assign all students to courses
for i in range(n):
    z_index.put([students_iq[i], students_iq[i], i])
    course_students_count[i] = 1
    
print(z_index.queue)
print(course_students_count)

# print(z_index.index(min(z_index, key = lambda t: t[0])))

for i in range(p):
    min_list = z_index.get()
    print(min_list[2] + 1, end=' ')
    course_students_count[min_list[2]] += 1
    z_index.put([(min_list[1] + monks_iq[i]) * course_students_count[min_list[2]], monks_iq[i], min_list[2]])
    print(z_index.queue)
```

```python
# using heapq

import heapq
# c, p, n = map(int, input().split())
# students_iq = list(map(int, input().split()))
# monks_iq = list(map(int, input().split()))

# c, p, n = 10, 10, 10
# students_iq = [2,8,5,1,10,5,9,9,3,5]
# monks_iq = [2,5,1,8,6,5,1,10,1,6]

c, p, n = 5, 4, 4
students_iq = [2, 8, 5, 1]
monks_iq = [9, 10, 5, 1]

# create a list of number of courses
course_students_count = [0]*c
# first item in each list is the z_index and second item is last_student's iq and third item is course number
z_index = []

# assign all students to courses
for i in range(n):
    z_index.append([students_iq[i], i, students_iq[i]])
    course_students_count[i] = 1
    
for i in range(n, c):
    z_index.append([0, i, 0])
    
print(z_index)
print(monks_iq)
heapq.heapify(z_index)
print(z_index)

print(course_students_count)
print()

# print(z_index.index(min(z_index, key = lambda t: t[0])))

for i in range(p):
    min_list = z_index[0]
    print(min_list[1] + 1, end=' ')
    course_students_count[min_list[1]] += 1
    new_z_index = (min_list[2] + monks_iq[i]) * course_students_count[min_list[1]]
#     z_index[0][1] = monks_iq[i]
    heapq.heapreplace(z_index, [new_z_index, min_list[1], monks_iq[i]])
    print(z_index)
#     heapq.heapify(z_index)
    print(z_index)
    print()
```
