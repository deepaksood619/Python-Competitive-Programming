```python
class Node:
    def __init__(self, data):
        self.prev = None
        self.next = None
        self.data = data
        
    def __str__(self):
        return '{}'.format(self.data)

class DLL:
    def __init__(self):
        self.curr_size = 0
        self.head = None
        self.tail = None
        
    def insert_head(self, data):
        # return pointer where the value is inserted
#         print('data', data, self.curr_size)
        node = Node(data)
        
        if self.curr_size == 0:
            self.tail = node
            self.head = node
        else:
            node.next = self.head
            self.head.prev = node
            self.head = node
        
        self.curr_size += 1
        
        return node

    def delete_tail(self):
        # return val of last item deleted to delete it's reference in dictionary
        node = self.tail
        self.tail = self.tail.prev
        self.tail.next = None
        
        self.curr_size -= 1
        
        return node.data
    
    def bring_to_front(self, node):
        # move to front
        # if head
        if self.head == node:
            return
        
        # if tail
        if self.tail == node:
            self.tail = node.prev
            node.prev.next = None
            node.next = self.head
            self.head.prev = node
            self.head = node
            node.prev = None
            
            return
            
        node.prev.next = node.next
        node.next.prev = node.prev
        node.next = self.head
        self.head.prev = node
        self.head = node
        node.prev = None
    
    def __str__(self):
        lst = []
        curr_pointer = self.head
        
        while curr_pointer != None:
#             print('curr', curr_pointer, curr_pointer.next, curr_pointer.prev)
            lst.append(curr_pointer.data)
            curr_pointer = curr_pointer.next
            
        return '->'.join(str(x) for x in lst)
        
class LRU:
    def __init__(self, size):
        self.dll = DLL()
        self.dic = {}
        self.size = size
        
    def get(self, data):
        print()
        # check if data present in cache
        node = self.dic.get(data)
        
        if node:
            # bring the val to front of DLL
            self.dll.bring_to_front(node)
            
            print(data, 'hit')
        else:
#             print('else', data)
            if self.dll.curr_size == self.size:
                last_val = self.dll.delete_tail()
                del self.dic[last_val]
            
            self.dic[data] = self.dll.insert_head(data)
            
            print(data, 'miss')
            
        print(self)
        
    def __str__(self):
        return self.dll.__str__()

cache = LRU(size=3)
cache.get(1)
# print(cache)
cache.get(2)
cache.get(3)
# print(cache)
cache.get(4)
cache.get(5)
cache.get(6)
# print(cache)
cache.get(4)
# print(cache)
cache.get(4)
cache.get(6)
cache.get(6)
# print(cache)
```
